```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate

model = ChatOpenAI(
    model_name="gpt-3.5-turbo"
)
```

```python
writing_prompt = ChatPromptTemplate.from_template(
    template="""
    You are a Haiku's guru.  
    mainly write haikus about {language}.
    """
)

writing_chain = writing_prompt | model
```

```python
explaining_prompt = ChatPromptTemplate.from_template(
    template="""
    You are a great Haiku's explainer. 
    You read {haiku} and then explain it.    
    """
)

explaining_chain = explaining_prompt | model
```

```python
final_chain = writing_chain | explaining_chain

final_result = final_chain.invoke({"language": "Python"})

final_result.content
```
