```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts.few_shot import FewShotPromptTemplate
from langchain_core.callbacks import StreamingStdOutCallbackHandler
from langchain_core.prompts import PromptTemplate
from langchain_classic.memory import ConversationBufferMemory
from dotenv import load_dotenv

load_dotenv()

model = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0.1,
    streaming=True,
    callbacks=[
        StreamingStdOutCallbackHandler(),
    ],
)

examples = [
    {
        "question": "How can you express Top Gun as three emojis?",
        "answer": "ğŸ›©ï¸ğŸ‘¨â€âœˆï¸ğŸ”¥",
    },
    {
        "question": "How can you express The Godfather as three emojis?",
        "answer":  "ğŸ‘¨â€ğŸ‘¨â€ğŸ‘¦ğŸ”«ğŸ ",        
    },
    {
        "question": "How can you express Jaws as three emojis?",
        "answer": "ğŸ¦ˆğŸ–ï¸ğŸ˜±",
    },    
]

example_prompt = PromptTemplate.from_template("Human: {question}\nAI: {answer}")

memory = ConversationBufferMemory(return_messages=True)

prompt = FewShotPromptTemplate(
    example_prompt=example_prompt,
    examples=examples,
    suffix="Human: How can you express {movie} as three emojis?",
    input_variables=["movie"],
)

chain = prompt | model

def invoke_chain(movie):
    result = chain.invoke({"movie": movie})
    memory.save_context({"input": movie}, {"output": result.content})
    print(result)

invoke_chain("Inception")
invoke_chain("The Dark Knight")
memory.load_memory_variables({})
```
